# 1. 배열

## 알고리즘

- 무엇이 좋은 알고리즘인가?
  1. 정확성
  2. 작업량(얼마나 적은 연산)
  3. 메모리 사용량
  4. 단순성
  5. 최적성

- 시간 복잡도(Time Complexity) 
  - 실행되는 명령문의 개수를 계산
  - 빅-오(O) 표기법
    - 시간 복잡도 함수 중에서 가장 큰 영향력을 주는 n에 대한 항만을 표시(최고차항)
    - 계수(Coefficient)는 생략하여 표시

## 배열

- 정렬(Sort)
  - 버블 정렬(Bubble)
    - 인접한 두 개의 원소를 비교하며 자리를 계속 교환하는 방식
    - 시간복잡도: `O(n^2)`
    - 알고리즘 기법: 비교와 교환

  - 카운팅 정렬(Counting)
    - 항목들의 순서를 결정하기 위해 집합에 각 항목이 몇 개씩 있는지 세는 작업을 하여, 선형 시간에 정렬하는 효율적인 알고리즘(n이 비교적 작을 때만 가능)
    - 시간복잡도: `O(n+k)` / n은 리스트 길이, k는 정수의 최댓값
    - 알고리즘 기법: 비교환 방식

  - 선택 정렬(Selection)
    - 교환의 회수가 버블, 삽입정렬보다 작다
    - 시간복잡도: `O(n^2)`
    - 알고리즘 기법: 비교와 교환

  - 퀵 정렬(Quick)
    - 최악의 경우 O(n^2)이지만 평균적으로는 가장 빠르다.
    - 시간복잡도: `O(n log n)` ~ `O(n^2)`
    - 알고리즘 기법: 분할 정복

  - 삽입 정렬(Insertion)
    - n의 개수가 작을 때 효과적이다.
    - 시간복잡도: `O(n^2)`
    - 알고리즘 기법: 비교와 교환

  - 병합 정렬(Merge)
    - 연결리스트의 경우 가장 효율적인 방식
    - 시간복잡도: `O(n log n)`
    - 알고리즘 기법: 분할 정복
