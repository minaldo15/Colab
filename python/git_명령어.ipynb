# git_basic

- `git init`
  - git 레포 만들기(홈 dir에서는 만들지 않는다)
  - 숨김 파일 .git 에 버전 정보 저장됨
  - (master - 깃으로 관리되고 있는 상태)

- 사용자 등록 (최초 커밋 전 필요)
  - `git config --global user.name "name"`
  - `git config --global user.email "email@naver.com"`
  - `git config --global -l` : 설정 확인

- 커밋
  - `git add`
    - 커밋할 내용 일단 저장(staging area)
  
  - `git commit -m "message"`
    - staging area의 모든 내용을 깃 저장소에 업데이트 with 메세지


- 구조
  - 실제 폴더
    - 파일 저장되어 있음
  - 로컬 git
    - working directory
      - 커밋하면 여기에 저장되어 변경사항 추적
    - commits
      - 커밋한 기록과 메세지 저장됨
    - staging area
      - 커밋할 내용 임시저장

# git_status
- 상태 확인
  - `git status`
    - 변경사항 있는지 확인

  - `git log`
    - 현 저장소의 변경사항 모두 출력
    - `git log --oneline -(숫자)` : 주요 내용만 출력 (몇줄)
    - `git log --oneline --graph` : 합쳐진 경로를 표시

- 깃허브 연동
  - `git remote add origin https://github.com/minaldo15/git-prac.git`
    - 원격으로 저장소 연결
  - `git remote -v`
    - 원격으로 연결한 저장소 확인
  - `git push origin master`
  	- 업로드
  	- origin - 레포 닉네임 (origin = {url})

- 빈폴더 업로드
  - `touch .gitkeep` 으로 더미 파일 생성
  - 그리고 push

# git_clone
- `mkdir git_clone` (폴더 생성)
- `git clone {url}` 
  - 깃 복사
- `git pull origin master` 
  - 깃 가져오기

- `git commit --amend` 
  - 정보(메세지 등등) 수정하기
  - `i` - 수정 입력(insert모드)
  - `esc` - insert모드 종료
  - `:wq` - 저장하고 입력 종료
  - `:q` - 종료
  - `:q!` - 저장안하고 강제종료
  
-secret 파일 무시하기(숨기기)
  - `git init` 하자마자 (`git add`하기전에)
  - `touch .gitignore` - 만들기
  - `secret.txt`를 `.gitignore` 메모장에 입력
  - `{파일이름}/` 입력 -> {}디렉토리 안에 있는 모든 파일 무시
  - `git status` 확인하면 `secret.txt` 파일 트래킹되지 않음

# git merge
- 각각 다른 로컬에서 
  - 다른 파일 작업(커밋)한 경우
    - home에서 push한 것을 cafe에서 pull한 후 push할 수 있다.
  - 같은 파일 작업한 경우(내용은 다름)
    - `git commit` 후 `code .`으로 VScode에서 병합(home 기준, cafe기준, 둘 다)선택한 후 push

# git reset/reflog
- `git reset {커밋네임}` - 그 커밋 이후로 날림
  - `--hard` - 전체 다(폴더,파일) 날라감
  - `--mixed` - 파일남음(add 날라감)
  - `--soft` - 파일 남음(add .남은상태)

- `git reflog`  
  - 이전에 있던 커밋들 다 보여줌
  - 그리고 다시 `git reset {원하는 커밋네임}` 으로 다시 복구
